```python
wiki2 = pandas.read_html(wiki) #solo funciona con tablas. Punto clave
#len(wiki2) #list of 28 pandas DataFrame 
#type(wiki2) #list
#type(wiki2[0]) #dataframe
```

```python
#consolidar todos los datos de todas los dataframes
wikiWW = wiki2[0]
wikiAF = wiki2[2]
wikiAN = wiki2[6]
wikiAC = wiki2[8]
wikiAS = wiki2[9]
wikiOM = wiki2[13]
wikiIN = wiki2[15]
wikiAO = wiki2[17]
wikiSA = wiki2[19]
wikiEOC = wiki2[23]
wikiEOR = wiki2[25]
wikiOC = wiki2[27]
wikiAM = wiki2[4]
wikiAS = wiki2[11]
wikiEU = wiki2[21]

wiki3_df = pandas.concat([wikiWW, wikiAF, wikiAN, wikiAC, wikiAS, wikiOM, wikiIN, wikiAO, wikiSA, wikiEOC, wikiEOR, wikiOC, wikiAM, wikiAS, wikiEU], axis=0)
wiki3_df
```

```python
#Exportar hacia un archivo csv todo el dataframe duplicado
wiki3_df.to_csv('wiki3_df.csv', index=False)
```

```python
wiki3_df.rename(columns = {'Poblacin segn Citypopulation (2015)': 'Citypopulation 2015',
                       'Poblacin segn Citypopulation (2016)': 'Citypopulation 2016',
                       'Poblacin segn Demographia (2015)': 'Demographia 2015',
                       'Poblacin segn ONU (2015)': 'ONU 2015',
                       'Poblacin segn ltimo censo': 'Ultimo Censo',
                       'Poblacin segn ltimo censo oficial': 'Ultimo Censo Oficial',
                       'Posicin': 'Posicin en Tabla Inicial', 
                       'Poblacin segn Citypopulation[1]\u200b': 'Citypopulation Sin Fecha', 'Poblacin segn Demographia[2]\u200b': 'Demographia Sin Fecha', 'Poblacin segn ONU[3]\u200b': 'ONU Sin Fecha', 'Fecha y fuente': 'Fecha'},
             inplace=True)
wiki3_df.columns
```

```python
#Al existir mltiples fechas para una misma fuente, se consolidarn en nuevas columnas generadas:
#el criterio de seleccin ser seleccionar fecha ms reciente o valor disponible
#primer paso generar nuevas columnas
#wiki3_df['Citypopulation'] = ''
#wiki3_df['Demographia'] = ''
#wiki3_df['ONU'] = ''
#wiki3_df.columns
#se descarta esta opcin debido que los inputs de las nuevas columnas generadas no provienen de una sola columna (no es copiar y pegar)
#se debe realizar una evaluacin (if: existe numero o no estaba vaco) o unir las columnas de origen (Citypopulation 2015, 2016 y sin fecha)
```

```python
#primero validar qu columnas estn vacias
null_cols = wiki3_df.isnull().sum()
null_cols
#null_cols[null_cols > 0]
```

```python
#las columnas con el mayor numero de valores vacios se eliminarn finalmente existen
#Existen 9 columnas con diferentes fuentes sobre el censo en las ciudades
wiki3_df = wiki3_df.drop(['Citypopulation 2016', 'Citypopulation Sin Fecha', 'Demographia Sin Fecha', 'ONU Sin Fecha', 'Ultimo Censo Oficial'], axis=1)
null_cols = wiki3_df.isnull().sum()
null_cols
```

```python
#confirmar los tipos de valores que tienen las columnas para poder sumar
print(wiki3_df.dtypes)
```

```python
#Antes de convertir a numeros, se debe limpiar la data de los puntos(.) de lo contrario ocurir error en la conversin
wiki3_df['Citypopulation 2015']= wiki3_df['Citypopulation 2015'].str.replace('.', '').replace('NaN', '0').replace('---','').replace(' ','') 
wiki3_df['Demographia 2015']= wiki3_df['Demographia 2015'].str.replace('.', '').replace('NaN', '0').replace('---','').replace(' ','') 
wiki3_df['ONU 2015']= wiki3_df['ONU 2015'].str.replace('.', '').replace('NaN', '0').replace('---','').replace(' ','')
wiki3_df['Ultimo Censo']= wiki3_df['Ultimo Censo'].str.replace('.', '').replace('NaN', '0').replace('---','').replace(' ','')
```

```python
#Convertir todos los valores de estimaciones de poblacin a nmeros
wiki3_df["Citypopulation 2015"] = pandas.to_numeric(wiki3_df["Citypopulation 2015"], errors='coerce')
wiki3_df["Demographia 2015"] = pandas.to_numeric(wiki3_df["Demographia 2015"], errors='coerce')
wiki3_df["ONU 2015"] = pandas.to_numeric(wiki3_df["ONU 2015"], errors='coerce')
wiki3_df["Fecha"] = pandas.to_numeric(wiki3_df["Fecha"], errors='coerce')
wiki3_df["Ultimo Censo"] = pandas.to_numeric(wiki3_df["Ultimo Censo"], errors='coerce')
#wiki3_df = wiki3_df.astype({"Posicin en la Tabla Inicial": int})
print(wiki3_df.dtypes)
```

```python
#Eliminar los rows que contengan valores "0" o "NaN": Cmo debe resultar?
wiki3_df.dropna(subset=['Citypopulation 2015'], how = 'all', inplace = True)
wiki3_df
```

```python
#Empezar a trabajar en limpiar la data:
#1. Eliminar registros duplicados
before = len(wiki3_df)
wiki3_df = wiki3_df.drop_duplicates()
after = len(wiki3_df)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
#2. Cul es el censo ms antiguo?
print(wiki3_df[wiki3_df.Fecha== wiki3_df.Fecha.min()])
```
