```python
#Se importan las librerias a utilizar
import requests
import pandas as pd
```

```python
# Con esto se hace la petición a la API
response = requests.get ('https://www.balldontlie.io/api/v1/players?page=0').json()
jugadores = pd.DataFrame(response['data'])
```

```python
# Utilice un for para extraer los datos de la base de datos ya que por cada página se desplegaban 25 datos
for page in range(1,60):
    response = requests.get('https://www.balldontlie.io/api/v1/players?page=%i' % page).json()
    solucion = pd.DataFrame(response['data'])
    jugadores = pd.concat([jugadores,solucion], axis = 0)

```

```python
# Aquí se despliega todo el data set
lista_jugadores = jugadores
jugadores
```

```python
# Quise saber el tipo de datos de este dataframe
lista_jugadores.dtypes
```

```python
# Las columnas del DataFrame de la lista de jugadores
columnas_jugadores = jugadores.columns
columnas_jugadores
```

```python
# El número de datos nulos por columna
null_cols = lista_jugadores.isnull().sum()
null_cols[null_cols > 0]
```

```python
# Dado que las columnas tenían datos nulos en su totalidad la removí
drop_cols = list(null_cols[null_cols > 1000].index)
data = lista_jugadores.drop(drop_cols, axis=1)

data.head(5)
```

```python
#Reordene las columnas
column_order = ['id','first_name','last_name','position','team']

lista_jugadores = lista_jugadores[column_order]
lista_jugadores.head()
```

```python
# El DataFrame  de la lista de jugadores en la columna Team tenía en sus datos anidados en diccionario
equipo = lista_jugadores['team']
equipo.values
```

```python
# Para desanidar los datos de la columna team, utilicé el mñétodo 'json_normalize' para desanidar los datos
from pandas.io.json import json_normalize

df_equipo = pd.DataFrame([equipo.values])
df_equipo = json_normalize(equipo)

df_equipo
```

```python
# Reordenamiento de las columnas del nuevo DataFrame de equipo
column_order = ['id','full_name','name','abbreviation','city','conference','division']

df_equipo = df_equipo[column_order]
df_equipo.head()
```

```python
#df_cd = pd.concat([lista_jugadores, df_equipo], axis=1, sort=False)
#Con un for pude unir los dos DataFrame
df_equipo.columns = ['team_id','team_name','name','abbreviation','city','conference','division']
columnas = df_equipo.columns

for column in columnas:
    lista_jugadores[column] = df_equipo[column]

lista_jugadores
```

```python
# Con drop me deshice de columnas que no eran utiles
bye = lista_jugadores.drop(['id','team','team_id'], axis=1)
bye
```

```python
# Con 'pd.values_counts' saqué el número de equipos por ciudad 
numero_city = pd.value_counts(lista_jugadores['city'])
numero_city 
```

```python
# Con 'pd.values_counts' saqué el número de jugadores por equipo
numero_player = pd.value_counts(lista_jugadores['team_name'])
numero_player
```

```python
# Con 'pd.concat' saqué una tabla de todos los jugadores del equipo de Memphis
n_c = bye[bye['city']=='Memphis']
n_p = bye[bye['team_name']=='Memphis Grizzlies']

memphis = pd.concat([n_c,n_p], axis=0)

memphis.head(20)
```

```python
# Con 'drop_duplicates' me deshice de los valores duplicados que eran 26
before = len(bye)
fresh_and_clean = bye.drop_duplicates()
after = len(bye)
print('Number of duplicate records dropped: ', str(before - after))
```

```python
# Con 'str.contains' saque los datos de quienes conforman el equipo de Miami Heat
players = bye[bye['team_name'].str.contains(('Miami Heat'))]
players.head()
```

```python
# Con 'sort_values' ordené de manera ascendente el nombre de los equipos para una mejor visualización
fresh_and_clean.sort_values(by=['team_name'], ascending=True)
```

```python
# Con 'to_csv' exporté este data set a un archivo csv
fresh_and_clean.to_csv('API-Basketball-clean.csv', index=False)
```
