#  Project API

```python
import requests 
import pandas as pd
from time import sleep

```

## Data Extraction Function

```python
def get_data(url,headers):
    responses = []
    for page in range(1, 11):
        sleep(2)
        print(f"Requesting {url%page} to server")
        sleep(2)
        response = requests.request("GET", url%page, headers=headers)
        response = response.json()   
        responses.append(response['results'])
        print(f"Response extracted successfully")
        print(10*"--------")
#         responses.append(response)
    return responses
```

## Data Transformation Function

```python
def convert_csv(response):
    print("Preparing information to be ready in file")
    sleep(2)
    flat_data = [final_data for matrix in response for final_data in matrix]
    df = pd.DataFrame(flat_data)
    parse = df.to_csv("./data/sneakers.csv", index=False)
    print("File ready on data folder with name 'sneakers.csv'")
    print(10*"--------")
    sleep(2)
    return parse
```

## General Function

```python
def run():
    headers = {
	"X-RapidAPI-Host": "the-sneaker-database.p.rapidapi.com",
	"X-RapidAPI-Key":""
    }
    url = 'https://the-sneaker-database.p.rapidapi.com/sneakers?limit=100&page=%s'
    print("The information is requested from API")
    response = get_data(url,headers)
    df = convert_csv(response)    
    
if __name__ == "__main__":
    run()
```
