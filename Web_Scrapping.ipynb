```python
def quotes_parser(content):
    soup = BeautifulSoup(content,'lxml')
    quotes = soup.select("div.so-booktitle h3 a")
    author = soup.select("span.so-bookprice")
    fotos = soup.select('div.so-frontcover a[href] img[src]')
    fotos = [foto['src'] for foto in fotos]
    links=[]
    for foto in fotos:
        if foto.startswith("/"):
            foto = "https://www.elsotano.com" + foto
            links.append(foto)
    links.pop(0) #
    quotes_text= [el.text for el in quotes]
    quotoes_text= [quote.replace("\n","") for quote in quotes_text]
    author_text= [el.text for el in author]
    author_text=[author.replace("\n","") for author in author_text]
    #quote_author = zip(quotes_text,author_text) #aqui estaba links
    #quote_author=list(quote_author)
    for i in range(3):
        quotes_text.pop(0)
        author_text.pop(0)
    quote_author_links=zip(quotes_text, author_text, links)
    quote_author_links=list(quote_author_links)
    return quote_author_links



# Instantiate the IronhackSpider class
my_spider = web_scrapping_proyecto(URL_PATTERN, PAGES_TO_SCRAPE, content_parser=quotes_parser)

# Start scraping jobs
my_spider.kickstart()


```
