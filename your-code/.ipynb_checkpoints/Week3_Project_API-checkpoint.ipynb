```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#Data exploration

data_response = [] #lista de los responses, por región (son 17 regiones, 17 páginas)
for q in range (1,18):
    response = requests.get(f'https://api.carbonintensity.org.uk/regional/intensity/2019-05-15T12:00Z/2019-05-16T12:00Z/regionid/{q}')
    data_response.append(response)
        
region_json =[]
for i in range(17):
    region = data_response[i].json()
    region_json.append(region)
#region_json

df_0 = pd.DataFrame(region_json)
df_0.to_csv('raw_data.csv')
```

```python
#Data unraveled
lista = []
for page in range(1,18):
    k = page
    url = f'https://api.carbonintensity.org.uk/regional/intensity/2019-05-15T12:00Z/2019-05-16T12:00Z/regionid/{k}'
    response_1 = requests.get(url)
    response_1 = response_1.json()

    results = response_1['data']['data']
    region = response_1['data']['shortname']

    for r in results:
        #print(r)
        temp = []
        temp.append(region)
        for element in r:
            #print(element)
            #temp.append(region)
            if element == 'from' or element == 'to':
                temp.append(r[element])
                #print(r[element])
            elif element == 'intensity':
                temp.append(r[element]['forecast'])
                temp.append(r[element]['index'])
            elif element == 'generationmix':
                for e in r[element]:
                    temp.append(e['perc'])
        lista.append(temp)      

df = pd.DataFrame(lista)
df.columns = ['region','from','to','forecast','index','biomass','coal','imports','gas','nuclear','other','hydro','solar','wind']
df
```

```python
#Searching null values

null_values = df.isnull().sum()
null_values
```
