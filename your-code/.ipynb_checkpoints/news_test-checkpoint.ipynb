```python
def search_news(searchTerm,lang):
    from requests_xml import XMLSession
    session = XMLSession()
    url = f"https://news.google.com/rss/search?q={searchTerm}&hl={lang}"
    global respuesta
    respuesta = session.get(url)
    return respuesta
```

```python
search_news('mezcal', 'es')
```

```python
def getNoticias():
    titulos = respuesta.xml.xpath('//title')
    titulos = titulos[1:]
    
    titulosList = [titulo.text for titulo in titulos]
    
    links = respuesta.xml.xpath('//link')
    links = links[1:]   
    
    global linkList
    linkList = [link.text for link in links]
 
    global diccionarioDeNotas
    
    diccionarioDeNotas = dict(zip(titulosList,linkList))
    
    for k, v in diccionarioDeNotas.items():
        print("Titulo de Articulo: {0} \nEnlace: {1} \n".format(k,v))
```

```python
getNoticias()
```
