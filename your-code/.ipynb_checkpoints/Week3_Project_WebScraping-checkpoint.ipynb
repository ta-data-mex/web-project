```python
# Libraries required
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pandas.io.json import json_normalize
import re
```

```python
#def trails_parser(content):

# URL of the general web page 
url = 'https://www.alltrails.com/es/mexico'
html = requests.get(url).content
html_soup = BeautifulSoup(html,'html')
```

```python
# findall the data I want

#List of links of the trails places, exclusively from MX
data_links = html_soup.find_all('a',{'class':'item-link'})
links_list = [links['href'] for links in data_links if links['href'].startswith('/es/trail/mexico')]
links = [l.replace('/es','https://www.alltrails.com/es') for l in links_list]

#List of the names in México 
data_names = html_soup.select('.item-info h3')
names_list = [name.text for name in data_names]
names_list = names_list[0:24]

#List of the locations in México
data_zone = html_soup.select('.location-label a')
zone_list = [zone.text for zone in data_zone]
zone_list = zone_list[0:24]

#List of the level
data_level = html_soup.select('.difficulty-info span')
level_list = [level.text for level in data_level]
lista = [level for level in level_list if not re.findall(r'\(\d+\)',level)] #taking just the names
lista = [l for l in lista if l]
level = lista[0:24]

#List of number of reviews
data_reviews = html_soup.select('.review-count')
reviews_list = [reviews.text for reviews in data_reviews]
# para indicar en el dt sólo los dígitos re.findall(r'\d+',star)
reviews_list =reviews_list[0:24]

#List of stars: I tried different ways but none was successful. Need your support here. 
#I tried:

#data_stars = html_soup.select('.rating')
#for i,j in enumerate(data_stars):
    #print(i[j]['title'])
    #lista.append(j[i]['title'])

#data_stars = html_soup.select_one('.difficulty-info > span[title]').get('title')
#data_stars #This only extract one of the ratings

#lista= []
#for i in range(len(data_stars)):
#    lista.append(data_stars[i]['title'])
#lista

#data_stars = html_soup.select('.rating',attrs = 'title')
#data_stars

#In addition, I would like to have further support to extract additional info from the pages ;)
```

```python
#Now I will check the collected data so far.

df_trails = pd.DataFrame({'Name of the Trail': names_list,
                          'Location':zone_list,'Level':level,
                          'Reviews':reviews_list,'Links':links}) 
df_trails
```

```python
df_trails.to_csv('raw_data_webScraping.csv')
```

```python
#Data cleaning

#Taking out parentheses in the Reviews column
df_trails['Reviews'] = [i.replace('(','').replace(')','') for i in df_trails['Reviews']]
```

```python
#Making a link inside the df
def make_clickable(val):
    return '<a target="_blank" href="{}">{}</a>'.format(val,val)

df_trails.style.format({'Links':make_clickable})
```
