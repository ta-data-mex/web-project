Estuve buscando en distintas páginas del gobierno de México para encontra una API a la cual pedirle información. En este caso encontré un set de información que no servía a mis propósitos pero que revisé que tenía. 

```python
# En este código se hace la petición a la API.
request = requests.get('https://api.datos.gob.mx/v1/calidadAire?date-insert>2019-04-05&page=154')
# Se imprimé la respuesta a la petición para saber si es correcta (200)
print(request)
# Se transforma la información solicitada que viene en un json a un diccionario. 
calidad_aire = request.json()
# Se normaliza la información para ver cómo está configurada
calidad_aire = json_normalize(calidad_aire)
# Como contiene un diccionario con diccionarios y listas de diccionarios adentro, para entrar en el que me interesa 
# normalizo o entro en el diccionario "results" que parece es donde viene la información que me interesa. 
flat_calidad_aire = json_normalize(calidad_aire.results[0])
# Este diccionario contiene más diccionarios así que normalizo la información sobre "stations" para ver qué contiene. 
flat_calidad_aire = json_normalize(flat_calidad_aire.stations[0])
# Imprimo la normalización. 
flat_calidad_aire
```

Una vez resuelto el probleme del acento en las palabras decido buscar los valores de esta misma partícula sólo en la Ciudad de México. Esto es para disminuir el tamaño de la información. Si lo hiciera por todo el sistema de estaciones habría más de 200,000 datos. Cabe aclarar que esta entrada de la API sólo tiene datos del 2018. 

```python
# Importo algunas librerías que voy a necesitar para descargar toda la información 
from pathlib import Path
import json 
import urllib

# En este código el rango se refiere al número de páginas que voy a descargar. El número de páginas totales viene en el
# json. Aproximadamente son 30 páginas con información aunque se informe que son más, el resto de ellas están vacías. 
for page in range(1, 20):
    filename = f'page_{page}_sinaica_mediciones.json' # se formatea la cadena para que el nombre salga por cada número de página
    filename = 'mediciones_sinaica_json/'+filename # esta variable contiene la ruta parcial en donde se va a guardar el archivo
    
    if not Path(filename).is_file(): # se verifica que no exista el arcihvo en la ruta y se hace la petición a la API
        estaciones_request = client.makeCall('sinaica', {'pageSize':5000, 'page':page,'parametro':'PM2.5',
                                                         'state':urllib.parse.quote_plus('Ciudad de México')}) # con este código hacemos que Jupyter pase el acento
        with open(filename, 'w')as outfile:
            json.dump(estaciones_request, outfile) # se abre el archivo en la ruta y se guarda la petición en un archivo json. 
```

```python
# Se define una función donde se abrirár los archivos json guardados y se concatenarán en una sóla base de datos.  
def parse_mediciones_json(json_file):
    with open(json_file, 'r') as aux:
        results = json.load(aux)['results'] # se abre el archivo y se lee 
        
    pre_data = []
    for r in results: 
        aux = pd.DataFrame.from_dict(r, orient = 'index').T # se convierte a un data frame y se orienta de manera correcta
        pre_data.append(aux) # se coloca en una lista vacía 
        
    if len(pre_data) > 0: # si el archivo contiene algo entonces se concatena
        pre_data = pd.concat(pre_data, ignore_index = True)
        return pre_data
    else:
        print("NO DATA FOR", json_file) # si el archivo no tiene información se imprime que esto mismo. 
        return
```

```python
# Se genera una variable name con un format para acceder a cada archivo guardado en la ruta
name = f'mediciones_sinaica_json/page_%i_sinaica_mediciones.json'
# Se usa un ciclo for para abrir cada uno de lo archivos substituyendo el %i por cada número en el rango.  
aux_list = [name %n for n in range(1, 20)]
# Se utiliza la función antes creada para abrir cada uno de los archivos y colocarlos en una lista de todos los archivos
# ya transformados a data frames. 
mediciones = [parse_mediciones_json(f) for f in aux_list]
# Una vez se tiene este arreglo de data frames se concatenan para generar una base de datos total. 
mediciones = pd.concat(mediciones, ignore_index = True)
# Se obtiene una base de datos de 95,000 filas por 11 columnas. 
mediciones
```

```python
# Se guarda este nuevo data frame total a un archivo csv para su fácil manipulación y trato de los datos. 
mediciones.to_csv('mediciones_sinaica_csv/mediciones_sinaica_pm2.5.csv', index = False)
```
