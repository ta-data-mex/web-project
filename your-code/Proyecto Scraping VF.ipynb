```python
import re
import requests

import numpy as np
import pandas as pd

from bs4 import BeautifulSoup
```

```python
url = 'https://www.kavak.com/autos-CR-V/compra-de-autos'
```

```python
reponse = requests.get(url)
```

```python
reponse.status_code
```

```python
soup = BeautifulSoup(reponse.content,'html.parser')
```

```python
soup
```

```python
modelo = soup.select('div h2.car-name')
modelo
```

```python
modelo_1 = [element.text for element in modelo]
modelo_1
```

```python
modelo_clean = [i.strip() for i in modelo_1] 
modelo_clean
```

```python
ao_km = soup.select('div p.car-details')
ao_km
```

```python
ao_km_1 = [element.text for element in ao_km]
ao_km_1
```

```python
ao_km_clean = [i.strip() for i in ao_km_1] 
ao_km_clean
```

```python
costo = soup.select('div.payment-amounts.reverse')
costo
```

```python
costo_1 = [element.text for element in costo]
costo_1
```

```python
costo_clean = [i.split() for i in costo_1] 
costo_clean
```

```python
len(modelo_clean), len(ao_km_clean),len(costo_clean) 
```

```python
df_1 = pd.DataFrame(zip(modelo_clean, ao_km_clean),
                 columns=['Modelo', 'Ao'])
```

```python
df_1.head(36)
```

```python
df_2 = pd.DataFrame()
```

```python
df_2[['precio_or','precio_de']] = costo_clean
df_2
```

```python
df_final = pd.concat([df_1, df_2], axis=1)
df_final
```

```python
df_final.to_csv('Proyecto_Web_Scrapi.csv')
```
