<h1> INTERNATIONAL NEWS </h1>

This is a web parser to identify title, summary and links of the international breaking news as published by different media of the Ibero-american region:

Right now parses:



<li>El País (Spain)</li>
<li>La Vanguardia (Spain)</li>
<li>El Milenio (México)</li>
<li>Clarín (Argentina)</li>


```python
import pandas as pd
from bs4 import BeautifulSoup
import html5lib
import lxml
import requests
import re
```

```python
import re
```

```python
class InternationalNews:
    
    def __init__(self):
        
        self.el_pais_internacional = []
        self.el_clarin_internacional = []
        self.output = pd.DataFrame(columns=['Titular', 'Summary', 'Link', 'Source', 'Country'])
        self.lv_internacional = []
        self.milenio_int = []
    
    def get_el_pais(self):
        
        el_pais = requests.get('https://elpais.com/tag/c/15148420ba519668342b7a63149cad97').content
        el_pais_soup = BeautifulSoup(el_pais, 'html5lib')
        el_pais_soup = el_pais_soup.select('body')
        el_pais_soup = el_pais_soup[0]
        el_pais_soup1 = el_pais_soup.find_all('div', {'id':'contenedor'})
        el_pais_soup1 = el_pais_soup1[0]
        el_pais_soup1 = el_pais_soup1.find_all('main', {'class':'principal'})
        el_pais_soup1 = el_pais_soup1[0]
        el_pais_soup1 = el_pais_soup1.find_all('section', {'class':'bloque_automatico'})
        el_pais_soup1 = el_pais_soup1[0]
        el_pais_soup1 = el_pais_soup1.find_all('div', {'class':'bloque__interior'})
        el_pais_soup1 = el_pais_soup1[0]
        el_pais_soup1 = el_pais_soup1.find_all('div', {'class':'articulos__interior'})
        
        for i in range(len(el_pais_soup1)):
            l = []
            titular = el_pais_soup1[i].select('h2')
            titular = titular[0].text
            l.append(titular)
            subtitulo = el_pais_soup1[i].select('p', {'class':'articulo-entradilla'})
            subtitulo = subtitulo[0].text
            l.append(subtitulo)
            link = el_pais_soup1[i].a['href']
            l.append('https:'+link)
            l.append('El Pais')
            l.append('España')

            self.el_pais_internacional.append(l)

        return  self.el_pais_internacional
    
    def clarin(self):
        
        requests.get('https://www.clarin.com/mundo/').content
        clarin_soup = BeautifulSoup(clarin, 'html5lib')
        clarin_soup = clarin_soup.find_all('div', {'class':'content home'})
        clarin_soup = clarin_soup[0]
        clarin_soup = clarin_soup.find_all('div', {'class':'box-notas'})
        clarin_soup = clarin_soup[0]
        l = []
        nota1 = clarin_soup.find_all('div', {'class':'mt pull-left'})
        nota1 = nota1[0]
        nota1 = nota1.a
        link = nota1['href']
        l.append('https://www.clarin.com'+link)
        titular = nota1.select('h2')
        titular = titular[0]
        l.append(titular.text)
        summary = nota1.select('p')
        summary = summary[1].text
        l.append(summary)
        l.append('Clarin')
        l.append('Argentina')
        self.el_clarin_internacional.append(l)
        l = []
        nota2 = clarin_soup.find_all('div', {'class':'box col-lg-3 col-md-4 col-sm-6 col-xs-12 no-p border'})
        nota2 = nota2[0]
        nota2 = nota2.a
        link = nota2['href']
        title = nota2.select('figure')
        title = title[0]
        title = title.select('img')
        title = title[0]
        title = title['alt']
        l.append(title)
        l.append('-')
        l.append('https://www.clarin.com'+link)
        l.append('Clarín')
        l.append('Argentina')
        self.el_clarin_internacional.append(l)

        return self.el_clarin_internacional
        
        
    def lv(self):
        lv = requests.get('https://www.lavanguardia.com/internacional').content
        lv_soup = BeautifulSoup(lv,'html5lib')
        lv_soup = BeautifulSoup(lv,'html5lib')
        lv_body = lv_soup.select('body')
        lv_div1 = lv_body[0].select('section', {'class':"section-block breakingnews small"})
        lv_div1 = lv_div1[0].select('header', {'class':"story-header-title-link"})
        lv_div1 = [self.lv_internacional.append([lv_div1[i].a['title'], '-', lv_div1[i].a['href'], 'La Vanguardia', 'España']) for i in range(5)]

        return self.lv_internacional
    
    def milenio(self):
        
        milenio = requests.get('https://www.milenio.com/internacional').content
        milenio_soup = BeautifulSoup(milenio,'html5lib')
        body = milenio_soup.select('body')
        notas = milenio_soup.find_all('div',{'class':'contenedor-notas-block'})
        notas = notas[0].find_all('div', {'class':'title'})
        for i in range(5):
            l = []
            link = notas[i].a['href']
            titular = notas[i].text
            titular = re.sub('\s\s+', '', titular)
            l.append([titular, '-', link, 'Milenio', 'México'])
            self.milenio_int.append(l)

        self.milenio_int = [y for i in self.milenio_int for y in i] 
        return self.milenio_int
    
    def main(self):
        
        self.clarin()
        self.get_el_pais()
        self.lv()
        self.milenio()

        
        dfc = [i for i in self.el_pais_internacional]+[i for i in self.el_clarin_internacional] + [i for i in self.lv_internacional] + [i for i in self.milenio_int]
        self.output = pd.DataFrame(dfc, columns=['Titular', 'Summary', 'Link', 'Source', 'Country'])
        
        return self.output
```

```python
news = InternationalNews()
news.main()
```
