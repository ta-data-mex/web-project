```python
bookmarks_cont = BeautifulSoup(page_source, 'html')
links_to_book = bookmarks_cont.select('div.latest_book a[href]')
links = [links_to_book[i]['href'] for i in range(len(links_to_book))]
links
```

### Extraccion de informacion del libro

```python
url_base = 'https://bookmarks.reviews'
```

```python
book_link_list = [url_base+links[i] for i in range(len(links))]
book_link_list
```

```python
def book_info(link):
    res = requests.get(link)
    book_info = BeautifulSoup(res.content, 'html')

    print(f'accesing {link} information')
    title = book_info.select('div h1')
    title_clean = [ti.text for ti in title]

    author = book_info.select('div.book_detail_author span')
    author_clean = [au.text for au in author]

    rate_info = book_info.select('div.book_review_stats div')
    rate_clean = [rate.text for rate in rate_info]

    num_reviews = book_info.select('a span[itemprop="ratingCount"]')
    num_reviews_clean = [review.text for review in num_reviews]

    publisher = book_info.select('div.detail_container_right span')
    publisher_clean = [publ.text for publ in publisher]

    release_date = book_info.select('div[content]')
    release_date_clean = release_date[0]['content']

    category_info = book_info.select('a span[itemprop="genre"]')
    category_clean = [categ.text for categ in category_info]

    reviews = book_info.select('div a.bookmarks_detail_see_all_reviews')
    reviews_clean = url_base+reviews[0]['href'] 
    print(f'ending')
    return title_clean[0], author_clean[0], rate_clean[0], num_reviews_clean[0], publisher_clean[0], release_date_clean, category_clean[0], reviews_clean
```

```python
link_info = [book_info(link) for link in book_link_list]
```

```python
df = pd.DataFrame(link_info, columns=['Book_Title', 'Auhtor', 'Overall_Review', 'Num_reviews',
                                              'Publisher', 'Release_date', 'Category', 'Link_all_reviews'])
df.to_csv('bookmark_best_reviewed.csv', index=False)
df.head()
```
