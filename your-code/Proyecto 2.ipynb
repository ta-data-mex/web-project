## Web Scraping - quelibroleo.com 

Se hará un WEB Scraping de la página WEB quelibroleo.com, de la cual se extraerá la informacion de los nombres y autores de los mejores libros biográficaos y memorias que presentan en su página. 

```python
contenido = []       # Es una lista con la info de las primeras 8 paginas de la listas de las mejores biografias de la web  
datos_varios = []
datos_puntuaciones = []
datos_links = []
puntaje=[]

#Se hace el request para obtener en formato la lista la informacion de las primeras 8 paginas de la página WEB 

for i in range(1,6):
    response = requests.get(f"https://quelibroleo.com/mejores-genero/biografias-memorias?sort=&page={i}").content
    contenido.append(response)
```

```python
paginas_cargadas = len(contenido)
print(paginas_cargadas)

datos_pag = paginas_cargadas*20
print(datos_pag)
```

```python
#Generar la sopa y extraer datos para sacar los nombres de los libros y autores, puntaje:

for i in range(len(contenido)):
    soup = BeautifulSoup(contenido[i], "lxml")
    a = soup.select("div[class='col-lg-8 col-xs-12'] a")
    b = soup.select("div[class='estadisticas'] span")
    datos_varios.append(a)
    datos_puntuaciones.append(b)
    
```

```python
# Función para extraer los nombres de los libros:

def nombres_libros(cont,datos):
    lst_libros = []
    
    for i in range(cont): 
        for j in (range(0,40,4)):
           
            a1 = datos[i][j].text
            lst_libros.append(a1)
            
    return lst_libros


libros = nombres_libros(paginas_cargadas,datos_varios)
```

```python
#Funcion para extraer los nombres de los autores:

def nombres_autores(cont, datos):
    lst_autores = []
    
    for i in range(cont):
        for j in (range(1,40,4)):
            b = datos[i][j].text
            lst_autores.append(b)
            
    return lst_autores

autores = nombres_autores(paginas_cargadas,datos_varios)
```

```python
def puntaje_libro(cont,datos):
    lst_puntuaciones = []
    
    for i in range(cont):
        for j in range(10):
            cali = datos[i][j].text.split()
            lst_puntuaciones.append(cali)
    
    return lst_puntuaciones

puntaje = puntaje_libro(paginas_cargadas,datos_puntuaciones)
```

```python
df_libro_autor_puntaje = pd.DataFrame(zip(libros, autores, puntaje), columns=["Libro", "Autor","Puntuación"])
df_libro_autor_puntaje
```

```python
df_libro_autor_puntaje.to_csv("Cincuenta_recomendaciones_de_libros_bibiograficos.txt", sep="\t", encoding="utf-16")
```
