# API de Ruteo de INEGI
https://www.inegi.org.mx/servicios/Ruteo/Default.html

```python
import time
from datetime import datetime 

from github import Github
from IPython.display import Image

import requests 
import time 
import re
from pathlib import Path
import pandas as pd
```

```python
url="https://gaia.inegi.org.mx/sakbe_v3.1/buscadestino"

palabras=["restaurante", "hotel", "aeropuerto", "cascada", "hospital", "gobierno"]

estados=["aguascalientes", "baja california", "baja california sur", "campeche", "chiapas", "chihuahua", "coahuila", "colima", 
         "ciudad de mexico", "durango", "guanajuato", "guerrero", "hidalgo", "jalisco", "mexico", "michoacan", "morelos",
         "nayarit", "nuevo leon", "oaxaca", "puebla", "querétaro", "quintana roo", "san luis potosí", "sinaloa", "sonora"
         "tabasco", "tlaxcala", "tamaulipas", "veracruz", "yucatan", "zacatecas"]

info=[]
for lugar in palabras:
    for entidad in estados:
        parametros={'buscar': f'{lugar}, {entidad}', 'type': 'json', 'num': '1000',
                    'key': "token"}
        info.append(requests.post(url, data = parametros).json())
        time.sleep(5)

info
```

```python
Series=[]

for datos in info:
    serie=datos['data']
    for i in serie:
        Series.append(i)
        
len(Series)
df = pd.DataFrame(Series)
df.head()
filepath = Path('ruteo.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df.to_csv(filepath)
```
