Introducción:

El siguiente scraping se realiza de la página de developer de Riot Games, ya que al realizar el proyecto de API encontré que hacía falta información que no brindaba la API, sin embargo había una sección de la página en la que venían dos tablas con la información faltante.

Objetivo:

Relacionar los dos proyectos, para obtener una base clara sobre las últimas 100 partidas del jugador seleccionado.

Resultado:

Se obtuvieron mediando web scraping dos dataframes con los datos faltantes en el proyecto de API.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://developer.riotgames.com/game-constants.html'
LOL = requests.get(url).text
```

```python
LOLsoup = BeautifulSoup(LOL,'html5lib')
LOLsoup
```

```python
queue = LOLsoup.find_all("tr")
queue_list = [q.text.replace('  ','').split('\n') for q in queue][13:90]
q = [t for t in queue_list if t]
queue_df = pd.DataFrame(q, columns= q[0])
queue_df = queue_df.drop(0).drop('', axis =1)
queue_df.to_csv('queue webscraping.csv', index = False)
queue_df = queue_df[queue_df.Notes == '']
queue_df['Id'] = queue_df['Id'].astype(np.int64)
del queue_df['Notes']
queue_df = queue_df.rename(columns={'Id':'queue'})
queue_df.to_csv('queue final.csv', index = False)
```
