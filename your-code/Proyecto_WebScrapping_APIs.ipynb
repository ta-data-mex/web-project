# Proyecto API's

```python
response = requests.get('https://officeapi.dev/api/characters')
```

```python
response.json()
```

```python
the_office = pd.DataFrame(response.json())
the_office.head()
```

```python
the_office1 = pd.json_normalize(the_office['data'])
the_office1.head()
```

```python
Characters = the_office1[['firstname', 'lastname']]
Characters
```

```python
Characters.to_csv('Characters.csv')
```

```python
Nombres = list(the_office1['firstname'])
Nombres
```

```python
Apellidos = list(the_office1['lastname'])
Apellidos
```

```python
Nombres_completos = list(zip(Nombres, Apellidos))
Nombres_completos
```

```python
np.savetxt("Nombres_completos.csv", Nombres_completos, fmt='% s')
```

```python
episodios = requests.get('https://officeapi.dev/api/episodes')
```

```python
episodios.json()
```

```python
data_episodios = pd.DataFrame(episodios.json())
data_episodios.head()
```

```python
data_episodios1 = pd.json_normalize(data_episodios['data'])
data_episodios1.head()
```

```python
titulos = data_episodios1['title']
titulos
```

```python
titulos_alfabeticamente = sorted(titulos)
titulos_alfabeticamente
```

```python
np.savetxt("Titulos.csv", titulos_alfabeticamente, fmt='% s')
```
