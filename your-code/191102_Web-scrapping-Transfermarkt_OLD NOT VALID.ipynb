```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#URL con la tabla del valor de los equipos en LaLiga para la temporada 2019-2020
url = 'https://www.transfermarkt.com/bundesliga/startseite/wettbewerb/L1/plus/?saison_id=2018'
#Para accesar la página
headers = {"User-Agent":"Mozilla/5.0"}

#Parsear el contenido 
response = requests.get(url, headers=headers, verify=False)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#Instrucciones para obtener todos los elementos de la tabla
table = soup.find('table', {'class' : 'items'})
table
```

```python
#Encuentra todas las filas(tr) de la tabla.
row = table.find_all('tr')
#Para conocer el número de filas en la tabla
len(row)
```

```python
team = []
squad = []
age = []
foreigners = []
total_market_value = []
average_market_value = []
for x in row:
    try:
        col = x.find_all('td')
        team.append(col[2].text)
        squad.append(col[3].text)
        age.append(col[4].text)
        foreigners.append(col[5].text)
        total_market_value.append(col[6].text)
        average_market_value.append(col[7].text)
    except:
        pass
```

```python

df_bundesliga = pd.DataFrame({'Team': team[1:],'No. Players': squad[1:], 
                              'Avg. Age': age[1:], 'No. Foreigners': foreigners[1:],
                              'Total Value': total_market_value[1:], 
                              'Average value': average_market_value[1:]})
#Limpiar y cambiar nomenclatura
df_bundesliga['Avg. Age'] = df_bundesliga['Avg. Age'].str.replace(',', '.')
df_bundesliga['Total Value'] = df_bundesliga['Total Value'].str.replace(',', '.')
df_bundesliga['Average value'] = df_bundesliga['Average value'].str.replace(',', '.')
df_bundesliga
```
