```python
res_json_2=res_2.json()  #aqui se metio el json en una variable
res_json_2
```

```python
datos_2 = res_json_2['data']  #aqui se saco solo la data que es lo que nos interesa sacar
datos_2
```

```python
df_2 = pd.DataFrame(datos_2) #aqui se genero un dataframe
df_2
```

```python
df_2_drop=df_2.drop(['volume','adj_high','adj_low','adj_close','adj_open','adj_volume','split_factor','dividend'], axis=1) #aqui se eliminaron las columnas que no nos funcionan
df_2_drop
```

```python
df_2_ordenado=df_2_drop.sort_values(by=['symbol','date'])  #aqui se ordenaron los datos por accion y por fecha
df_2_ordenado
```

```python
res_3 = requests.get('http://api.marketstack.com/v1/eod?access_key=d8c17415fa92ad84b011013f35f0d125&symbols=TSLA,MSFT,AMZN&date_from=2021-09-01&date_to=2021-09-30')
#se va a sacar las acciones lo del mes de septiembre 2021 de las 3 acciones
```

```python
res_3
```

```python
res_json_3=res_3.json()  #aqui se metio el json en una variable
res_json_3
```

```python
datos_3 = res_json_3['data']  #aqui se saco solo la data que es lo que nos interesa sacar
datos_3
```

```python
df_3 = pd.DataFrame(datos_3) #aqui se genero un dataframe
df_3
```

```python
df_3_drop=df_3.drop(['volume','adj_high','adj_low','adj_close','adj_open','adj_volume','split_factor','dividend'], axis=1) #aqui se eliminaron las columnas que no nos funcionan
df_3_drop
```

```python
df_3_ordenado=df_3_drop.sort_values(by=['symbol','date'])  #aqui se ordenaron los datos por accion y por fecha
df_3_ordenado
```

```python
df_sep_oct = pd.concat([df_2_ordenado, df_3_ordenado], axis = 1)
df_sep_oct
```

```python
df_sep_oct.columns= ['Precio_apertura','Precio_mas_alto','Precio_mas_bajo','Precio_cierre','Accion','Indice','Fecha_Oct','Precio_apertura','Precio_mas_alto','Precio_mas_bajo','Precio_cierre','Accion','Indice','Fecha_Sep']
df_sep_oct.columns
```
