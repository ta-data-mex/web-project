```python
#Importamos todorxio
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#La url de la API es:
url = 'https://futar.bkk.hu/api/query/v1/ws/otp/api/where/bicycle-rental.json?key=apaiary-test&version=3&appVersion=apiary-1.0&includeReferences=true'
response = requests.get(url) 
results = response.json()
results
```

```python
#Rental dataframe, nos despliega la información muy truculenta
rental_df = pd.DataFrame(results)
rental_df
```

```python
#Generamos el primer csv con esa información
rental_csv = rental_df.to_csv('rental.csv', index='False')
```

```python
#Rental dataframe modificado ahora sí para que nos de toda la info que necesitamos
total_df = pd.DataFrame(results['data']['list'])
total_df
```

```python
#Al final generamos un nuevo dataframe sólo con las columnas que nos interesan 
final_df = pd.DataFrame(total_df[['name', 'bikes','code','id']])
final_df
```

```python
#Exportamos nuestro csv final 
total_csv = final_df.to_csv('rental_modified.csv', index='False')
```
