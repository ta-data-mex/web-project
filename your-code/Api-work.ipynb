```python
import requests
import pandas as pd
import json
```

```python
#el api tiene 214 paginas con 10 personajes en cada una. Hice un for loop para sacar todos los api y 
#meterlos a la lista fire
fire = []
for i in range(1, 214):
    url = f'https://www.anapioficeandfire.com/api/characters?page={i}&pageSize=10'
    soiaf = requests.get(url)
    song = soiaf.json()
    fire.append(song)
```

```python
#asi queda la lista de personajes
fire
```

```python
#la lista fire era una lista de listas entonces hice una nueva lista (ice) donde 
#estaria cada elemento de las listas dentro de fire
ice = [z for y in fire for z in y]
ice
```

```python
#teniendo ice lo convertí en un dataframe
ice_and_fire = pd.DataFrame(ice)
ice_and_fire
```

```python
#el orden de las columnas era confuso asi que cambie el orden
cols = ['name', 'aliases', 'titles', 'born', 'died', 'gender', 'culture', 'father', 'mother', 'books']
ice_and_fire = ice_and_fire[cols]
ice_and_fire
```

```python
#cada valor vacio dentro de la columna titles estaba en una lista con un valor vacio
#asi que hice un for para meter el string que queria y saque el valor vacio
for x in ice_and_fire['titles']:
    if '' in x:
        x.append('Has no title')
        x.pop(0)
ice_and_fire
```

```python
#por error volvi a correr la celda y entonces tuve que borrar los valores repetidos de 'has no title'
for x in ice_and_fire['titles']:
    if 'Has no title' in x:
        while len(x) >1:
            x.pop(0)      
ice_and_fire
```

```python
#hice lo mismo con la columna de aliases, elimine las listas con valores vacios y lo sustitui por 'has no known aliases'
for x in ice_and_fire['aliases']:
    if '' in x:
        x.append('Has no known aliases')
        x.pop(0)
ice_and_fire
```

```python
#la columna de books tenia cada libro en el que habia aparecido el personaje asi que hice una lista con el numero 
#libros en el que aparecia cada personaje
b_apperences = [len(x) for x in ice_and_fire['books']]
b_apperences

    
```

```python
#añadi esa lista como una nueva columna
ice_and_fire['Books Apperences'] = b_apperences
```

```python
#me di cuenta de que el personaje de arya no tenia libros en los que apareciera y
#como eso es un error le di el mismo valor que otro personaje que sale en los mismos libros
#2037 es el indice del otro personaje y 134 el de arya
ice_and_fire['books'][2037] = ice_and_fire['books'][134]
```

```python
#actualice el dataframe ordenando los valores de mayor a menor por numero de libros en los que aparecen
ice_and_fire = ice_and_fire.sort_values('Books Apperences', ascending = False)
```

```python
ice_and_fire
```

```python
#queria analizar la diferencia del genero de los personajes asi que lo ordene por gender
gender = ice_and_fire.groupby('gender').count()
```

```python
#saca el porcentaje de cada genero
perc_female_character = gender['name']['Female'] / len(ice_and_fire['name']) * 100
perc_male_character = gender['name']['Male'] / len(ice_and_fire['name']) * 100
print(perc_female_character, perc_male_character)
porcentajes = [perc_female_character, perc_male_character]
```

```python
#otro dataframe donde se ve la diferencia en el genero de los personajes
gender['Porcentaje por genero'] = porcentajes
columnas = ['Books Apperences', 'Porcentaje por genero']
gender = gender[columnas]
gender
```

```python
#hice lo mismo con la culture para saber de donde viene cada personaje
culture = ice_and_fire.groupby('culture').count()
```

```python
#oredene el dataframe de mayor a menor por personajes de cada cultura
culture = culture.sort_values('Books Apperences', ascending = False)
culture
```

```python
#saque el porcentaje de cada cultura como origen de personajes
culture_porc = [(x / len(ice_and_fire['name']) * 100) for x in culture['Books Apperences']]
culture_porc
```

```python
culture['Porcentaje por cultura'] = culture_porc
```

```python
#junte las columnas de porcentaje y numero de personaje por cultura para dejar otro dataframe
culture_cols = ['Books Apperences', 'Porcentaje por cultura']
culture = culture[culture_cols]
culture
```

```python
#cambie el nombre del indice de los personajes que no se sabe su cultura de origen
culture.rename(index = {"": "Unknown Culture"}, inplace = True)
```
