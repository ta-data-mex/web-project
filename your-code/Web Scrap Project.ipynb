```python
#Importamos las librerias a usar

import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
#traemos el url de la pagina que vamos a usar (FreeCompany en FFXIV)

url = 'https://na.finalfantasyxiv.com/lodestone/freecompany/9228579323923932444/member/?page=1'
html = requests.get(url).content
links = BeautifulSoup(html, 'html5lib')
```

```python
#transformamos en Data Frame para poder hacerlo un csv
linksdf = pd.DataFrame(links)
```

```python
#Traemos los elementos que vamos a usar, en este caso sera el nombre del personaje, su mundo/server y el rango en la guild

player = links.find_all('p',{'class' : 'entry__name'})
```

```python
world = links.find_all('p',{'class' : 'entry__world'})
```

```python
rank = links.find_all('ul',{"class" : "entry__freecompany__info"})
```

```python
#Transformamos nuestros datos en Strings para la aplicacion de Reguklar Expresions

playerstr = str(player)
worldstr = str(world)
rankstr = str(rank)
```

```python
#Aqui limpiamos nuestros datos de los elementos que vamos a ocupar, usamos regular expresions para quitar los tags
#Usamos Split para regresarlos a una lista y limpiamos los corchetes con replace

cleanp = re.sub(r'<(.*?)>','', playerstr)
cleanp = cleanp.split(',')
cleanp = [i.replace('[','').replace(']','') for i in cleanp]
```

```python
cleanw = re.sub(r'<(.*?)>','', worldstr)
cleanw = cleanw.replace('\xa0','')
cleanw = cleanw.split(',')
cleanw = [i.replace('[','').replace(']','') for i in cleanw]
```

```python
cleanr = re.sub(r'<(.*?)>','', rankstr)
cleanr = re.sub(r'\d','',cleanr)
cleanr = cleanr.split(',')
cleanr = [i.replace('[','').replace(']','') for i in cleanr]
```

```python
#convertimos nuestros datos en DataFrames para poder unirlos en uno solo

p_df = pd.DataFrame(cleanp)
w_df = pd.DataFrame(cleanw)
r_df = pd.DataFrame(cleanr)
```

```python
#unimos nuestros DataFrames usando concat
#renombramos nuestras columnas 

fc_df = pd.concat([p_df, w_df], axis=1)
fc_df2 = pd.concat([fc_df, r_df], axis=1)
columns = ['Player_Name','World/Server','Rank']
fc_df2.columns = columns
```

```python
#Usamos iloc para eliminar los elementos apartir del 51 ya que no son miembros de la Free Company

fc_df2 = fc_df2.iloc[:50]
```
