```python
# Step 1: Load the webpage containing the data.
url = 'https://www.transfermarkt.es/liga-mx-clausura/marktwerte/wettbewerb/MEX1/pos//detailpos/0/altersklasse/alle/plus/1'
html = requests.get(url).content
headers = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url, headers=headers).content
soup = BeautifulSoup(response, 'html')
print(soup.prettify())
```

```python
players_table = soup.select('table.items tr')
```

```python
players_table = [[e for e in p if e!= '\n'] for p in players_table]
```

```python
players_table1 = [[e.text if e.text else e.img['alt'] for e in p] for p in players_table]
```

```python
players_table_final = [e for e in players_table1[:] if len(e) > 2]
```

```python
name_position = soup.find_all("table", {"class": "inline-table"})
name_position_list = [i.text.replace('\n', '') for i in name_position]
```

```python
nested_position = [i.split()[2:] for i in name_position_list]
```

```python
position_final = [" ".join(nested_position[i]) for i in range(len(nested_position))]
```

```python
players = soup.find_all("a", {"class": "spielprofil_tooltip"})
players_list = [i.text for i in players]
players_list_f = players_list[1:]
```

```python
df = pd.DataFrame(players_table_final[1:], columns=players_table_final[0])
df.insert(1, 'Name', players_list_f)
df.insert(2, 'Posición', position_final)
```

```python
final_dataframe = df.drop("Jugadores", axis=1)
final_dataframe.head()
```

```python
final_dataframe.to_excel('transfermkt_ligamx.xlsx', index=False)
```
