```python
#todod los pasos anteriores los meti en esta funcion para 
#que solo dando el input del url de la pgina de estadisticas te lo ordene y limpie en un data frame
def df_nfl(url):
    data_att = requests.get(url).content
    soup_att = BeautifulSoup(data_att, "html")
    text_att = soup_att.find_all(['td'])
    text_att = [x.text for x in text_att]
    txt_attempts = [(text_att[x:x+4]) for x in range(0, len(text_att), 4)]
    att_df = pd.DataFrame(txt_attempts)
    df_att = att_df.drop(3, axis = 1)
    new = df_att[2].str.split("-", n = 1, expand = True)
    df_att['Debut year'] = new[0]
    df_att['Last year'] = new[1]
    years = [(int(x) - int(new[0][i])) for i, x in enumerate(new[1])]
    years = pd.DataFrame(years)
    df_att['Years played'] = years
    df_att = df_att.drop(2, axis = 1)
    return df_att
```

```python
#use la funcin para la informacin de los pases completos y renombro las primeras columnas igual que el primer
#dataframe para poder unirlos ms adelante
url2 = "https://www.pro-football-reference.com/leaders/pass_cmp_career.htm"
df_com = df_nfl(url2)
df_com = df_com.rename(columns={0:'Player', 1: 'Completions'})
df_com
```

```python
url3 = 'https://www.pro-football-reference.com/leaders/pass_yds_career.htm'
df_yds = df_nfl(url3)
df_yds = df_yds.rename(columns={0:'Player', 1: 'Yards'})
df_yds
```

```python
url4 = 'https://www.pro-football-reference.com/leaders/pass_td_career.htm'
df_td = df_nfl(url4)
df_td = df_td.rename(columns={0:'Player', 1: 'TD'})
df_td
```

```python
url5 = 'https://www.pro-football-reference.com/leaders/pass_int_career.htm'
df_int = df_nfl(url5)
df_int = df_int.rename(columns={0:'Player', 1: 'Interceptions'})
df_int
```

```python

url6 = 'https://www.pro-football-reference.com/leaders/pass_rating_career.htm'
df_rtg = df_nfl(url6)
df_rtg = df_rtg.rename(columns={0:'Player', 1: 'Passer Rating'})
df_rtg
```
