```python
import warnings
warnings.simplefilter('ignore')

import time
import re
import sys
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd

def get_page_source(url):
    '''Esta es una función que extrae el codigo fuente de las estadisticas
    de los jugadores de la UEFA Champions League, al ser una pagina dinámica se
    utiliza el método de Selenium para extraer los datos de esta.'''
    driver=webdriver.Chrome(r'C:\Program Files\Chromedriver\chromedriver.exe')
    path=r'C:\Program Files\Chromedriver\chromedriver.exe'
    url="https://es.uefa.com/uefachampionsleague/statistics/players/"
    driver=webdriver.Chrome(path)
    driver.get(url)
    time.sleep(3)
    ActionChains(driver).send_keys(Keys.END).perform()
    time.sleep(5)
    ActionChains(driver).send_keys(Keys.END).perform()
    time.sleep(5)
    ActionChains(driver).send_keys(Keys.END).perform()
    time.sleep(5)
    ActionChains(driver).send_keys(Keys.END).perform()
    time.sleep(5)
    page_source=driver.page_source
    driver.quit()
    return page_source

def get_player(page_source):
    '''Esta es una función que procesa los datos que se extraen de la función get_page_source
    para realizar la sopa y buscar los nombres de los jugadores.'''
    word_selection=BeautifulSoup(page_source, 'lxml')
    nombre_jugador=word_selection.select('span[slot="primary"]')
    print(nombre_jugador)
    player=[]
    for jugador in nombre_jugador:
        player.append(jugador.text)
    return player

def get_equipo(page_source):
    '''Esta es una función que procesa los datos que se extraen de la función get_page_source
    para realizar la sopa y buscar los nombres de los equipos.'''
    equipo_jugador=word_selection.select('span[slot="secondary"]')
    print(equipo_jugador)
    all_equipos=[]
    for equip in equipo_jugador:
        all_equipos.append(equip)
    return all_equipos

def get_csv():
    from pathlib import Path
    df1 = pd.DataFrame(all_equipos)
    df2 = pd.DataFrame(player)
    df_final = pd.concat([df1,df2], axis=1)
    df_final
    filepath = Path('Champions_League2.csv')  
    filepath.parent.mkdir(parents=True, exist_ok=True)  
    final=df_final.to_csv(filepath)
    return final

get_csv()
```
