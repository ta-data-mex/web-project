## Periodico El Financiero Bloomberg ##

Se busca traer el ultimo precio de la divisas

```python
import re
```

```python
import requests
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.elfinanciero.com.mx/mercados/divisas/'
```

```python
response = requests.get(url)
```

```python
response
```

```python
html = response.content
```

```python
html
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
soup
```

```python
soup.find_all('td')
```

```python
links = soup.select('td a[class="decoration_none hover_color_black color_black text_left font_bold atag"]')  #De aqui voy a sacar los links y los nombres
```

```python
links
```

```python
links[0].text  #Aqui se probo si esta sacando el texto de la los links
```

```python
links[0]['href'] #Aqui se probo si esta sacando la liga de los links
```

```python
len(links)  #Aqui se probo la cantidad de links que hay 
```

```python
precio = soup.select('td[class="padding-xs-all text_vertical-align_middle text_right"]') # De aqui se esta sacando el precio de la divisa
```

```python
precio
```

```python
precio[0].text #Aqui se probo si esta saliendo el precio, pero esta sacando mas datos que NO son el precio
```

```python
nombre = [nomb.text for nomb in links]  #Aqui se hizo una list comprehension para sacar los nombres de las divisas ya que estan en texto
```

```python
nombre
```

```python
len(nombre) #Aqui se esta viendo si me da la misma cantidad de los links
```

```python
url_base = 'https://www.elfinanciero.com.mx'  #Se esta usando la url base ya que en la variable de links, el link viene incompleto.
```

```python
links_de_divisas= [url_base+links[i]['href'] for i in range(len(links))]  #Aqui se hizo una list comprehension para poder sacar los links completos
```

```python
links_de_divisas
```

```python
len(links_de_divisas) #Aqui se esta viendo si da la misma cantidad que los nombres
```

```python
precio_de_divisa = [prec.text for prec in precio] 
```

```python
type(precio)
```

```python
precio_de_divisa
```

```python
len(precio_de_divisa)
```

```python
precio_2 = [precio_de_divisa[pr] for pr in range(len(precio_de_divisa))if pr%2==1] #[element for element in iterable if condition]
```

```python
precio_2
```

```python
precio_3=[element for element in precio_2 if re.search(r'\d.*\.\d.*', element)]
```

```python
precio_3
```

```python
df = pd.DataFrame(zip(nombre, links_de_divisas, precio_3), columns=['Nombre', 'Liga','Ultimo_Precio'])
```

```python
df
```

```python
df.to_csv('Precios_divisas.csv')
```
